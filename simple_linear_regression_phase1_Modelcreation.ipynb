```python
# pandas is a library function used for working with structured data 
#in which the data are organized as tables such as spreadsheets or database. 
import pandas as pd
```

```python
# loading salary database which is a csv file to a term 'dataset'using pandas 
dataset = pd.read_csv("Salary_Data.csv")
dataset
```

```python
# Separating the input and output from the dataset. Selecting the input variable as 'YearsExperience' column from the dataset
independent=dataset[["YearsExperience"]]
independent
```

```python
#selecting the output variable as 'Salary' column from the dataset
dependent= dataset[["Salary"]]
dependent
```

```python
#using machine learning module,train_test_split function is essential for splitting data into two sets: one for training the model, and the other for testing it. 
from sklearn.model_selection import train_test_split
```

```python
# train_test_split function helps in randomly splitting the dataset into these two subsets, while ensuring 
#that can control the proportion of data in each subset
X_train,X_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# this is the procedure to do a project in machine learning particulary in simple linear regression model.
from sklearn.linear_model import LinearRegression
```

```python
# The LinearRegression() class is used to create a regressor object that will learn the relationship between the features (X_train) 
# and the target variable (y_train).fit() method computes the coefficients (slopes) of the linear regression model 
# that minimize the error between the predicted values and the actual values in the training data.
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# calculating the weight coefficient in y = wx+b
weight=regressor.coef_
weight
```

```python
# calculating the bias intercept
bias=regressor.intercept_
bias
```

```python
# the trained model which was fitted on X_train and y_train to predict the target values for the test data X_test.
# this is used to cross-check to predict the values
y_pred=regressor.predict(X_test)
```

```python
# starts the Evaluation metrics using R2 score which indicates how well the predictions (y_pred) match the actual target values (y_test)
from sklearn.metrics import r2_score
```

```python
# if the r2 score is near to '1.0'then it indicates perfect prediction and if it is near to '0.0' the model is no better than predicting 
# the mean of the target variable for all instances.,
r_score=r2_score(y_test,y_pred)
r_score
```

```python
# to save the model
import pickle
```

```python
# storing a name in sav file
filename="finalized_model_linear.sav"
```

```python
# saving the model,your trained regressor model will be saved to a file named 'Finalized_model_linear' using wb-write binary
pickle.dump(regressor,open(filename,'wb'))
```

```python
# again loading the same model in the read mode
loaded_model=pickle.load(open("finalized_model_linear.sav", 'rb'))
```
